```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium

print('Folium installed and imported!')
```

```python
latitude =  37.335480
longitude = -121.893028
```

```python
SJ_map = folium.Map(location=[latitude, longitude], zoom_start=12)

SJ_map
```

```python
limit = 1000
df = df.iloc[0:limit, :]
```

```python
# instantiate a feature group for the incidents in the dataframe
crashes = folium.map.FeatureGroup()

# loop through the 1000 incidents and add each to the crashes feature group
for lat, lng, in zip(df.Latitude, df.Longitude):
    crashes.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
SJ_map.add_child(crashes)
```

## Limit the number of crashes for a better view 

```python

crashes = folium.map.FeatureGroup()


for lat, lng, in zip(df.Latitude, df.Longitude):
    crashes.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=3, # define how big the circle markers should be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# adding pop-up text to each marker on the map

latitudes = list(df.Latitude)
longitudes = list(df.Longitude)
labels = list(df.CollisionType)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(SJ_map)    
    
# add crashes to map
SJ_map.add_child(crashes)
```
